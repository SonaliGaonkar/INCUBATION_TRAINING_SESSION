```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('data.csv')
```

```python
dataset
```

```python
datasetlist = ['ENGLISH','SCIENCE','MATHS','HINDI','HISTORY']
dataset['TOTAL_MARKS'] = dataset[datasetlist].sum(axis=1)
dataset
```

```python
dataset['PERCENTAGE']=(dataset['TOTAL_MARKS']/500)*100
dataset
```

```python
dataset.loc[dataset['PERCENTAGE']>=90, 'GRADES']='A+'
dataset.loc[((dataset['PERCENTAGE']>=80) & (dataset['PERCENTAGE']<90)), 'GRADES']='A'
dataset.loc[((dataset['PERCENTAGE']>=70) & (dataset['PERCENTAGE']<80)), 'GRADES']='B+'
dataset.loc[((dataset['PERCENTAGE']>=60) & (dataset['PERCENTAGE']<70)), 'GRADES']='B'
dataset.loc[((dataset['PERCENTAGE']>=50) & (dataset['PERCENTAGE']<60)), 'GRADES']='C+'
dataset.loc[((dataset['PERCENTAGE']>=35) & (dataset['PERCENTAGE']<50)), 'GRADES']='C'
dataset.loc[dataset['PERCENTAGE']<35, 'GRADES']='F'

dataset
```
